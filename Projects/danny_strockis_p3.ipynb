```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.mixture import GMM
from matplotlib.colors import LogNorm
```

Load feature names.

```python
feature_names = []
with open('../Data/mushroom.map') as fmap:
    for line in fmap:
        [index, name, junk] = line.split()
        feature_names.append(name)

print 'Loaded feature names:', len(feature_names)
```

Load data. The data is sparse in the input file, but there aren't too many features, so we'll use a dense representation, which is supported by all sklearn objects.

```python
X, Y = [], []

with open('../Data/mushroom.data') as fdata:
    for line in fdata:
        items = line.split()
        Y.append(int(items.pop(0)))
        x = np.zeros(len(feature_names))
        for item in items:
            feature = int(item.split(':')[0])
            x[feature] = 1
        X.append(x)

# Convert these lists to numpy arrays.
X = np.array(X)
Y = np.array(Y)

# Split into train and test data.
train_data, train_labels = X[:7000], Y[:7000]
test_data, test_labels = X[7000:], Y[7000:]

# Check that the shapes look right.
print train_data.shape, test_data.shape
```

```python
def P1():
### STUDENT START ###

    for k in range(1,50):
        pca = PCA(n_components=k)
        pca.fit(train_data)
        print 'k=' + str(k) + ': ' + str(sum(pca.explained_variance_ratio_))

### STUDENT END ###

P1()
```

(3) Run KMeans with [1,16] clusters over the 2d projected data. Mark each centroid cluster and plot a circle that goes through the most distant point assigned to each cluster.

```python
from matplotlib.patches import Circle

def P3():
### STUDENT START ###

    # Reduce dimensions
    pca = PCA(n_components=2)
    reduced_train_data = pca.fit_transform(train_data)

    for k in range(1,17):

        print 'k=' + str(k) + '\n-------------------------'
        
        # Train KMeans, store cluster assignments and distance to each cluster for each data point
        model = KMeans(n_clusters=k).fit(reduced_train_data)
        cluster_assignments = model.predict(reduced_train_data)
        cluster_distances = model.transform(reduced_train_data)

        # compute the longest distance to a data point for each cluster
        cluster_max_distances = np.zeros(k)
        for i in range(len(cluster_distances)):
            if cluster_distances[i][cluster_assignments[i]] > cluster_max_distances[cluster_assignments[i]]:
                cluster_max_distances[cluster_assignments[i]] = cluster_distances[i][cluster_assignments[i]]

        # plot everything
        fig,ax = plt.subplots(1)
        ax.set_aspect('equal')
        ax.scatter(reduced_train_data[:,0], reduced_train_data[:,1], c=train_labels, cmap=matplotlib.colors.ListedColormap(['r','b']))
        ax.plot(model.cluster_centers_[:,0], model.cluster_centers_[:,1], '+')
        for i in range(len(model.cluster_centers_)):
            ax.add_patch(Circle((model.cluster_centers_[i][0], model.cluster_centers_[i][1]), cluster_max_distances[i], ec='black', fc='none'))

        plt.show()
### STUDENT END ###

P3()
```

(5) Fit two 4-component full covariance GMMs, one for the positive examples and one for the negative examples in your 2d projected data. Predict the test examples by choosing the label for which the model gives a larger probability (use GMM.score). What is the accuracy?

```python
def P5():
### STUDENT START ###

    # Reduce dimensions
    pca = PCA(n_components=2)
    reduced_train_data = pca.fit_transform(train_data)
    reduced_test_data = pca.transform(test_data)
    
    # Split data in two
    positive_reduced_train_data = reduced_train_data[np.where(train_labels == 1)]
    negative_reduced_train_data = reduced_train_data[np.where(train_labels == 0)]
    
    # Train models
    positive_model = GaussianMixture(n_components=4, covariance_type='full')
    negative_model = GaussianMixture(n_components=4, covariance_type='full')
    positive_model.fit(positive_reduced_train_data)
    negative_model.fit(negative_reduced_train_data)
    
    # Compute per sample scores
    positive_scores = positive_model.score_samples(reduced_test_data)
    negative_scores = negative_model.score_samples(reduced_test_data)
    
    # Compare probabilities
    predictions = (positive_scores > negative_scores).astype('int')
    
    # Score accuracy and print
    print metrics.accuracy_score(test_labels, predictions)
    
### STUDENT END ###

P5()
```

(6) Ideally, we'd like a model that gives the best accuracy with the fewest parameters. Run a series of experiments to find the model that gives the best accuracy with no more than 50 parameters. For example, with 3 PCA components and 2-component diagonal covariance GMMs, you'd have:

( (3 mean vector + 3 covariance matrix) x 2 components ) x 2 classes = 24 parameters

You should vary the number of PCA components, the number of GMM components, and the covariance type.

```python
def P6():
### STUDENT START ###

    highest_accuracy = 0
    highest_accuracy_text = ''

    # Vary covariances
    for cov in ['full', 'tied', 'diag', 'spherical']:

        # Vary number of dimensions
        for j in range(1,12):

            # Vary number of clusters
            for k in range(1,12):
                
                # Use definition of covariance types
                cov_factor = 0
                if cov == 'full':
                    cov_factor = 2
                    num_components = k
                    num_dimensions = j
                if cov == 'tied':
                    cov_factor = 2
                    num_components = 1
                    num_dimensions = j
                elif cov == 'diag':
                    cov_factor = 1
                    num_components = k
                    num_dimensions = j
                elif cov == 'spherical':
                    cov_factor = 0
                    num_components = k
                    num_dimensions = j

                # Calculate number of parameters
                num_params = ((num_dimensions + num_dimensions ** cov_factor) * num_components)* 2

                # Cap the number of parameters at 50
                if num_params >= 50:
                    continue

                # Reduce dimensions
                pca = PCA(n_components=j)
                reduced_train_data = pca.fit_transform(train_data)
                reduced_test_data = pca.transform(test_data)

                # Split data in two
                positive_reduced_train_data = reduced_train_data[np.where(train_labels == 1)]
                negative_reduced_train_data = reduced_train_data[np.where(train_labels == 0)]

                # Train models
                positive_model = GaussianMixture(n_components=k, covariance_type=cov)
                negative_model = GaussianMixture(n_components=k, covariance_type=cov)
                positive_model.fit(positive_reduced_train_data)
                negative_model.fit(negative_reduced_train_data)

                # Compute per sample scores
                positive_scores = positive_model.score_samples(reduced_test_data)
                negative_scores = negative_model.score_samples(reduced_test_data)

                # Compare probabilities
                predictions = (positive_scores > negative_scores).astype('int')

                # Score accuracy and print
                accuracy = metrics.accuracy_score(test_labels, predictions)
                
                if accuracy >= highest_accuracy:
                    highest_accuracy = accuracy
                    highest_accuracy_text = 'PCA components = {0}, GMM components = {1}; Cov type = {2}: Num params = {3}, Accuracy = {4}'.format(
                        str(j),
                        str(k),
                        str(cov),
                        str(num_params),
                        str(accuracy)
                    )
                    
    print highest_accuracy_text

### STUDENT END ###

P6()
```
